```python
#ЗАДАНИЕ 1
#Дан объект в 2D пространстве

import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# При помощи линейных отображений:
# Уменьшить объект в два раза

# Матрица для уменьшения объекта
A2 = np.array([[0.5, 0,    0],
       [0,  0.5,    0],
       [ 0,  0,    0.5]])
# умножаем диагональную матрицу и уменьшаем объект в 2 раза
A3=A@A2

x_small = A3[:,0]
y_small = A3[:,1]
x = A[:,0]
y = A[:,1]
plt.plot(x_small, y_small)
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Повернуть на 130 градусов
A_orto=np.array([[np.cos(130), -np.sin(130), 0],[np.sin(130), np.cos(130), 0],[0, 0, 1]]) # ортогональная матрица
A4=A@A_orto # умножаем основную матрицу на ортогональную - поворот против часовой на 130 градусов относительно 0
x1 = A4[:,0]
y1 = A4[:,1]
plt.plot(x1, y1)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])
A_mirror=np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
A5 = A@A_mirror # Умножаем на симматричную матрицу
x2 = A5[:,0]
y2 = A5[:,1]
plt.plot(x2, y2)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# ЗАДАНИЕ 2
# Найдите спектральное разложение матрицы:
# m_spectr = vec_m*m_diag*vec_m_tr
m = np.array([[1, 2],
              [2, 3]])
print(m)

num, vec_m = np.linalg.eigh(m) #получаем собственные вектора и значения матрицы
```

```python
vec_m # собственные векторы
```

```python
num # собственные значения
```

```python
m_diag = np.diag(num) # создаем диагональную матрицу на  основе собственных значений
m_diag
```

```python
vec_m_tr = np.transpose(vec_m) # создаем транспонированную версию матрицы собственных векторов
vec_m_tr
# матрица является симметричной, потому транспонированная версия будет равна базовой
```

```python
# Итоговый результат. Перемножаем все между собой
m_spectr = vec_m@m_diag@vec_m_tr
m_spectr
# Спектральное разложение матрицы и ее начальное значения получились идентичными
```
